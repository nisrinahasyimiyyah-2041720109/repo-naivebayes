```python
# Import library yang dibutuhkan
import matplotlib.pyplot as plt
import numpy as np

# Masukkan data manual
labels = ['Stopword Pengecualian\nNegasi - Tanpa Stemming', 
          'Stopword Pengecualian\nNegasi - Stemming', 
          'Stopword Lengkap\n - Tanpa Stemming', 
          'Stopword Lengkap\n - Stemming']
accuracy = [0.94, 0.93, 0.94, 0.93]
precision = [0.94, 0.93, 0.95, 0.93]
recall = [0.94, 0.93, 0.94, 0.93]
f1_score = [0.94, 0.92, 0.94, 0.92]

# Lebar bar
bar_width = 0.15

# Posisi bar pada sumbu x dengan jarak antar kelompok
r1 = np.arange(len(labels))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]

# Buat figure dan axis
fig, ax = plt.subplots(figsize=(14, 10))

# Warna-warna yang lebih menarik
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# Buat bar chart
bars1 = ax.bar(r1, accuracy, color=colors[0], width=bar_width, label='Accuracy')
bars2 = ax.bar(r2, precision, color=colors[1], width=bar_width, label='Precision')
bars3 = ax.bar(r3, recall, color=colors[2], width=bar_width, label='Recall')
bars4 = ax.bar(r4, f1_score, color=colors[3], width=bar_width, label='F1-Score')

# Tambahkan angka di atas setiap bar dengan penyesuaian posisi
def add_values_on_bars(bars):
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, yval + 0.005, f'{yval:.2f}', ha='center', va='bottom', fontsize=14)

add_values_on_bars(bars1)
add_values_on_bars(bars2)
add_values_on_bars(bars3)
add_values_on_bars(bars4)

# Tambahkan judul
ax.set_title('Perbandingan Akurasi Training Pelabelan SentiStrength_ID - Naive Bayes', fontsize=20)

# Hilangkan label sumbu y dan angka pada sumbu y
ax.set_ylabel('')
ax.set_yticks([])

# Tambahkan legenda dan geser ke kanan
ax.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=14)

# Set posisi sumbu x dan label
ax.set_xticks([r + 1.5*bar_width for r in r1])
ax.set_xticklabels(labels, fontsize=14)

# Set layout agar tidak overlap
plt.tight_layout()

# Tampilkan bar chart
plt.show()
```
